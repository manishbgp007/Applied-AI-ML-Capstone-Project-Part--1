##Task 2: Null Value Analysis**

* We computed missing counts and percentages for each column.
* Columns with >20% nulls were flagged for special handling.
* Numeric columns with <20% nulls were imputed using the median.
  * Justification: The median is less sensitive to skewness and outliers compared to the mean, making it a more robust choice.## **

```python
#Task 2:- Null Value Analysis
# Count and percentage of nulls
null_counts = df.isnull().sum()
null_percent = (df.isnull().sum() / df.shape[0]) * 100
null_table = pd.DataFrame({"Null Count": null_counts, "Null %": null_percent})
print(null_table)

# Columns >20% nulls
print("Columns >20% nulls:", null_table[null_table["Null %"] > 20].index.tolist())

# Fill numeric columns <20% nulls with median
for col in df.select_dtypes(include=['float64','int64']).columns:
    if null_percent[col] < 20:
        df[col].fillna(df[col].median(), inplace=True)
```

#### **Task 3: Duplicate Detection**

* We identified duplicate rows using .duplicated().sum().
* Duplicates were removed with .drop_duplicates().
* The number of rows removed was reported.
* Null percentages were rechecked to confirm no unintended changes.

```python
#Task 3 — Duplicate Detection
# Count duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check shape after removal
print("New shape:", df.shape)
```

#### **Task 4: Data Type Correction**

* We corrected misclassified columns:
  * Converted numeric columns stored as object to proper numeric types.
  * Converted repetitive string columns to category dtype.
  * Compared memory usage before and after conversion — categorical conversion reduced memory footprint significantly.

```python
# Task 4 — Data Type Correction

import pandas as pd

# 1. Convert numeric columns stored as object to numeric
# This loop will automatically find all object-type columns and try to convert them
for col in df.select_dtypes(include='object').columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Example: if you know a specific column should be numeric, you can do:
# df['HouseAge'] = pd.to_numeric(df['HouseAge'], errors='coerce')

# 2. Convert repetitive string column to category
# Replace 'Region' with the actual column name in your dataset that contains repeated strings
if 'Region' in df.columns:
    df['Region'] = df['Region'].astype('category')

# 3. Memory usage before and after conversion
print("Memory usage after conversion:", df.memory_usage(deep=True).sum())
```

#### **Task 5: Descriptive Statistics & Skewness**

* We generated descriptive statistics with .describe() and computed skewness for each numeric column.
* Identified the column with the highest absolute skewness.

* Interpretation:
  * Positive skew → long right tail, mean > median.
  * Negative skew → long left tail, mean < median.
  * Consequence: Median is preferred for imputation in skewed distributions.

```python
#Task 5 — Descriptive Statistics & Skewness
# Descriptive stats
print(df.describe())

# Skewness
skew_vals = df.skew(numeric_only=True)
print(skew_vals)

# Column with highest absolute skewness
max_skew_col = skew_vals.abs().idxmax()
print("Most skewed column:", max_skew_col)
```

#### **Task 6: Outlier Detection (IQR)**

* For two numeric columns, we calculated Q1, Q3, and IQR.

  * Counted rows outside the bounds (outliers).
  * Outliers were documented but not removed.
  * Strategy: Retain them for now, consider capping or advanced handling in Part 2

```python
#Task 6 — Outlier Detection (IQR)
for col in df.select_dtypes(include=['float64','int64']).columns[:2]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {outliers.shape[0]} outliers")
```

#### **Task 9b: Spearman vs Pearson Correlation**

* We computed Spearman rank correlation and compared it to Pearson.
* Identified three column pairs with largest |Spearman − Pearson| differences.
* Interpreted whether relationships were monotonic but non-linear or approximately linear.
* Chose Spearman for feature selection guidance in Part 2, as it captures monotonic relationships missed by Pearson.

```python
#Task 9b — Spearman vs Pearson
pearson = df.corr(numeric_only=True)
spearman = df.corr(method='spearman', numeric_only=True)

diff = (spearman - pearson).abs()
print("Difference table:\n", diff)

# Top 3 differences
print(diff.unstack().sort_values(ascending=False)[1:4])
```

#### **Task 9c: Grouped Aggregation**

* We grouped one categorical column by one numeric column and computed mean, std, and count.
* Reported group with highest mean and highest std.
* Discussed implications of high within-group variance for predictive modeling.
* Calculated ratio of highest to lowest mean — a large ratio suggests predictive signal in the categorical feature.

```python
# Task 9C — Grouped Aggregation

import pandas as pd

# Step 1: Create a categorical column by binning latitude into 5 regions
df['LatBin'] = pd.cut(df['latitude'], bins=5)

# Step 2: Choose a numeric column to aggregate (median_house_value is the target)
num_col = 'median_house_value'
cat_col = 'LatBin'

# Step 3: Perform grouped aggregation
agg_result = df.groupby(cat_col)[num_col].agg(['mean','std','count'])
print("Grouped Aggregation Result:\n", agg_result)

# Step 4: Highest mean and std groups
print("Highest mean group:", agg_result['mean'].idxmax())
print("Highest std group:", agg_result['std'].idxmax())

# Step 5: Ratio of highest to lowest mean
ratio = agg_result['mean'].max() / agg_result['mean'].min()
print("Mean ratio:", ratio)
```
