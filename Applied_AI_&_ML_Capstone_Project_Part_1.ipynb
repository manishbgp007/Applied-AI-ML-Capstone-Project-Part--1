```python
from google.colab import files
import pandas as pd

# Upload the CSV file
# Load into DataFrame
df = pd.read_csv("/content/pd.read.csv")
#Task 1 — Inspect Dataset
# First 5 rows
print(df.head())

# Column data types
print(df.dtypes)

# Shape of DataFrame
print("Shape:", df.shape)
```

```python
#Task 2:- Null Value Analysis
# Count and percentage of nulls
null_counts = df.isnull().sum()
null_percent = (df.isnull().sum() / df.shape[0]) * 100
null_table = pd.DataFrame({"Null Count": null_counts, "Null %": null_percent})
print(null_table)

# Columns >20% nulls
print("Columns >20% nulls:", null_table[null_table["Null %"] > 20].index.tolist())

# Fill numeric columns <20% nulls with median
for col in df.select_dtypes(include=['float64','int64']).columns:
    if null_percent[col] < 20:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
#Task 3 — Duplicate Detection
# Count duplicates
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check shape after removal
print("New shape:", df.shape)
```

```python
# Task 4 — Data Type Correction

import pandas as pd

# 1. Convert numeric columns stored as object to numeric
# This loop will automatically find all object-type columns and try to convert them
for col in df.select_dtypes(include='object').columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Example: if you know a specific column should be numeric, you can do:
# df['HouseAge'] = pd.to_numeric(df['HouseAge'], errors='coerce')

# 2. Convert repetitive string column to category
# Replace 'Region' with the actual column name in your dataset that contains repeated strings
if 'Region' in df.columns:
    df['Region'] = df['Region'].astype('category')

# 3. Memory usage before and after conversion
print("Memory usage after conversion:", df.memory_usage(deep=True).sum())
```

```python
#Task 5 — Descriptive Statistics & Skewness
# Descriptive stats
print(df.describe())

# Skewness
skew_vals = df.skew(numeric_only=True)
print(skew_vals)

# Column with highest absolute skewness
max_skew_col = skew_vals.abs().idxmax()
print("Most skewed column:", max_skew_col)
```

```python
#Task 6 — Outlier Detection (IQR)
for col in df.select_dtypes(include=['float64','int64']).columns[:2]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
# Task 7 — Visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Check available columns
print("Columns in dataset:", df.columns)

# Step 2: Select numeric and categorical columns dynamically
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
categorical_cols = df.select_dtypes(include=['object','category']).columns

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# Line plot: use the first numeric column
plt.plot(df[numeric_cols[0]])
plt.title(f"Line Plot of {numeric_cols[0]}")
plt.xlabel("Index")
plt.ylabel(numeric_cols[0])
plt.show()

# Bar chart: mean of first numeric column grouped by first categorical column
if len(categorical_cols) > 0:
    df.groupby(categorical_cols[0])[numeric_cols[0]].mean().plot.bar()
    plt.title(f"Bar Chart: Mean {numeric_cols[0]} by {categorical_cols[0]}")
    plt.xlabel(categorical_cols[0])
    plt.ylabel(f"Mean {numeric_cols[0]}")
    plt.show()

# Histogram: most skewed numeric column
max_skew_col = df[numeric_cols].skew().abs().idxmax()
sns.histplot(df[max_skew_col], bins=20)
plt.title(f"Histogram of {max_skew_col}")
plt
```

```python
#Task 8 — Correlation Heatmap
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Highest absolute correlation pair
corr_unstack = corr_matrix.abs().unstack().sort_values(ascending=False)
print("Strongest correlation pair:", corr_unstack[1])  # skip self-correlation
```

```python
#Task 9a — Imputation Strategy Comparison
for col in skew_vals.abs().nlargest(2).index:
    print(col, "Mean:", df[col].mean(), "Median:", df[col].median())
    df[col].fillna(df[col].median(), inplace=True)

# Confirm no nulls
print(df.isnull().sum())
```

```python
#Task 9b — Spearman vs Pearson
pearson = df.corr(numeric_only=True)
spearman = df.corr(method='spearman', numeric_only=True)

diff = (spearman - pearson).abs()
print("Difference table:\n", diff)

# Top 3 differences
print(diff.unstack().sort_values(ascending=False)[1:4])
```

```python
# Task 9C — Grouped Aggregation

import pandas as pd

# Step 1: Create a categorical column by binning latitude into 5 regions
df['LatBin'] = pd.cut(df['latitude'], bins=5)

# Step 2: Choose a numeric column to aggregate (median_house_value is the target)
num_col = 'median_house_value'
cat_col = 'LatBin'

# Step 3: Perform grouped aggregation
agg_result = df.groupby(cat_col)[num_col].agg(['mean','std','count'])
print("Grouped Aggregation Result:\n", agg_result)

# Step 4: Highest mean and std groups
print("Highest mean group:", agg_result['mean'].idxmax())
print("Highest std group:", agg_result['std'].idxmax())

# Step 5: Ratio of highest to lowest mean
ratio = agg_result['mean'].max() / agg_result['mean'].min()
print("Mean ratio:", ratio)
```
